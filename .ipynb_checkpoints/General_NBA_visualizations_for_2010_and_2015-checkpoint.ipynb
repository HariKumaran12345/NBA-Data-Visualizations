```python
#imports
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('dataset/nbaallelo.csv')
dataset['point_differential'] = dataset['pts'] - dataset['opp_pts'] #create new column
```

```python
nba2010 = dataset[dataset.year_id == 2010]
nba2015 = dataset[dataset.year_id == 2015]
```

```python
nba2010.head()
```

```python
nba2015.head()
```

```python
nuggets2010Pts = nba2010[nba2010.fran_id == 'Nuggets']['pts']
celtics2010Pts = nba2010[nba2010.fran_id == 'Celtics']['pts']
```

```python
plt.clf()
plt.hist(nuggets2010Pts, alpha = 0.8,label = 'Nuggets')
plt.hist(celtics2010Pts, alpha = 0.8,label = 'Celtics')
plt.legend()
plt.title("2010 Season")
plt.show()
```

```python
nuggets2015Pts = nba2015[nba2015.fran_id == 'Nuggets']['pts']
celtics2015Pts = nba2015[nba2015.fran_id == 'Celtics']['pts']
plt.clf()
plt.hist(nuggets2015Pts, alpha = 0.8,label = 'Nuggets')
plt.hist(celtics2015Pts, alpha = 0.8,label = 'Celtics')
plt.legend()
plt.title("2015 Season")
plt.show()
```

```python
plt.clf()

plt.figure(figsize=(14, 8))

# Create the boxplot
sns.boxplot(data=nba2015, x='fran_id', y='pts', width=0.5, dodge=True)

plt.xticks(rotation=45)

plt.show()
```

```python
locationResult = pd.crosstab(nba2015.game_result, nba2015.game_location)
print(locationResult)
locationResultProp = locationResult/len(nba2015)
```

```python
locationResultProp = locationResult/len(nba2015)
print(locationResultProp)
```

```python
chi2, pval, dof, expected = chi2_contingency(locationResult)
print(expected)
print(chi2)
```

```python
diffCov = np.cov(nba2015.point_differential, nba2015.forecast)
forecastCorrelation = pearsonr(nba2015.forecast, nba2015.point_differential)
print(forecastCorrelation)
```

```python
plt.clf()
plt.scatter('point_differential', 'forecast', data = nba2015)
plt.xlabel('Point Differential')
plt.ylabel('Forecasted Win Probability')
plt.show()
```
